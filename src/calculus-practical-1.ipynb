    $$

**D. Prove the ff. (5 pts)**

9. Using PMI, prove that

    $$
    1^3 + 2 ^3 + 3^3 + \dots + n^3 = \left[ \frac{1}{2} (n (n + 1)) \right]^2
    $$

    for all $n \in \mathbb{N}$.

    **Answer:**

    Assume $n = 1$:

    $$
    1^3 = 1, \left[ \frac{1(1 + 1)}{2} \right]^2 = \left[ \frac{1 \cdot 2}{2} \right]^2 = 1^2 = 1\\
    $$

    The formula holds for $n = 1$.

    Assume that the formula holds for some $n = k$:

    $$
    1^3 + 2^3 + 3^3 + \dots + k^3  = \left[ \frac{k(k + 1)}{2} \right]^2
    $$

    We need to prove that the formula also holds for $n = k + 1$:

    $$
    1^3 + 2^3 + 3^3 + \dots + (k + 1)^3 = \left[ \frac{(k + 1)(k + 2)}{2} \right]^2
    $$

    Using the induction hypothesis:

    $$
    \begin{align}
    1^3 + 2^3 + 3^3 + \dots + k^3 + (k + 1)^3 & = \left[ \frac{k(k + 1)}{2} \right]^2 + (k + 1)^3 \\
    & = \frac{k^2(k + 1)^2}{4} + {k + 1}^3 \\
    & = (k + 1)^2 \left( \frac{k^2}{4} + (k + 1) \right) \\
    & = (k + 1)^2 \left( \frac{k^2 + 4(k + 1)}{4} \right) \\
    & = (k + 1)^2 \left( \frac{k^2 + 4k + 4}{4} \right) \\
    & = (k + 1)^2 \times \frac{(k + 2)^2}{4} \\
    & = \left[ \frac{(k + 1)(k + 2)}{2} \right]^2
    \end{align}
    $$
    
    **Therefore**, by using the pinciple of mathematical induction, the formula

    $$
    1^3 + 2 ^3 + 3^3 + \dots + n^3 = \left[ \frac{1}{2} (n (n + 1)) \right]^2
    $$

    holds for all $n \in \mathbb{N}$.
