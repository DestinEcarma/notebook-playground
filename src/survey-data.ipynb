# Towards a Fairer and More Prosperous Future – Prioritizing Our Development Focus

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

##################
# Synthetic Data #
##################

np.random.seed(69)

SIZE = 45

df = pd.DataFrame()
df["id"] = list(range(1, SIZE + 1))

for i in range(1, 7):
    # 2 options question
    df[f"2OQ {i}"] = np.clip(np.random.normal(2, 0.15, SIZE).astype(int), 1, 2)

    # Rate 1-10 question
    df[f"RQ {i}"] = np.clip(
        np.random.normal(7.8, 2.5, SIZE).astype(int), 1, 10
    )

####################
# Helper Functions #
####################


def plot_2options(df: pd.DataFrame, question: int) -> None:
    col = f"2OQ {question}"

    counts = df[col].value_counts().sort_index()

    labels = ["SDG 8", "SDG 10"]

    plt.figure(figsize=(8, 5))

    plt.bar(labels, counts, color=["#4c72b0", "#dd8452"], edgecolor="black")

    for i, count in enumerate(counts):
        plt.text(
            i, count + 0.3, str(count), ha="center", va="bottom", fontsize=10
        )

    plt.ylabel("Respondents")
    plt.title(f"Distribution of 2 Options Question #{question}")

    plt.grid(axis="y", linestyle="--", alpha=0.7)

    plt.show()


def plot_rating(df: pd.DataFrame, question: int) -> None:
    col = f"RQ {question}"

    plt.figure(figsize=(8, 5))

    counts, bins, patches = plt.hist(
        df[col],
        bins=range(1, 12),
        align="left",
        rwidth=0.8,
        color="#4c72b0",
        edgecolor="black",
    )

    for count, patch in zip(counts, patches):
        height = patch.get_height()
        plt.text(
            patch.get_x() + patch.get_width() / 2,
            height + 0.3,
            str(int(count)),
            ha="center",
            va="bottom",
            fontsize=10,
        )

    plt.xlabel("Rating")
    plt.ylabel("Respondents")
    plt.title(f"Distribution of Rating Question #{question}")

    plt.xticks(range(1, 11))

    plt.grid(axis="y", linestyle="--", alpha=0.7)

    plt.show()
```

## Section 1: 2 Option
